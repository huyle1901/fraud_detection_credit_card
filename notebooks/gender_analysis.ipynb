```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Read file CSV 
file_path = 'C:/Users/Admin/Desktop/fraud_detection_project/data/processed/clean_credit_card_transactions.csv'
df = pd.read_csv(file_path)
```

```python
# Overview of data
df.head()
```

```python
# Analyze according to gender
# 1. Which gender have most transaction ?

# Group by 'gender' and count the transaction
gender_counts = df.groupby("gender").size().reset_index(name='counts')

gender_labels = ['Female', 'Male']

colors = sns.color_palette("pastel")

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.pie(gender_counts['counts'], labels=None, autopct='%1.1f%%', startangle=90,colors=colors)

# Set labels and title
plt.title('Number of Transactions by Gender')

plt.legend(gender_labels, title="Gender", loc="best")



# Save the graph
image_path = r"C:\Users\Admin\Desktop\fraud_detection_project\reports\figures\gender\number_of_transactions_by_gender.png"
plt.savefig(image_path)

# Display the plot
plt.show()
```

```python
# 2. Which gender have most total transaction amout of money ?

# Group by 'gender' and sum the 'amt' for each category
gender_amount = df.groupby("gender")["amt"].sum().reset_index()

gender_labels = ['Female', 'Male']

colors = sns.color_palette("Set2")

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.pie(gender_amount['amt'], labels=None, autopct='%1.1f%%', startangle=90,colors=colors)

# Set labels and title
plt.title('Total Transaction Amount by Gender')

plt.legend(gender_labels, title="Gender", loc="best")

# Save the graph
image_path = r"C:\Users\Admin\Desktop\fraud_detection_project\reports\figures\gender\total_transaction_amount_by_gender.png"
plt.savefig(image_path)

# Display the plot
plt.show()
```
